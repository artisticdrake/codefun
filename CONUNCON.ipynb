```python
import pandas as pd
df=pd.read_csv(r'mycsv.csv')
df.head(10)
```

```python
def uncon():
  for i in df:
   print(i)
  col=int(input("Select the index of the column required\n"))
  nl=[]
  for i in range(len(df)):
    nl.append(df.iloc[i,col])
  nl=list(set(nl))
  print(nl)
  val=int(input("Enter the index of required item\n\n"))
  c=0
  for i in range(len(df)):
    if(df.iloc[i,col]==nl[val]):
      c+=1
  pa=c/len(df)
  print("The unconditional probability of ",nl[val]," is: ",pa)
```

```python
def con():
  nl=[]
  nl2=[]
  for i in df:
   print(i)
  col=int(input("Select the index of the column required for event A\n"))
  
  for i in range(len(df)):
    nl.append(df.iloc[i,col])
  nl=list(set(nl))
  print(nl)
  val=int(input("Enter the index of required item for event A\n\n"))
  for i in df:
   print(i)
  col2=int(input("Select the index of the column required for event B\n"))
  
  for i in range(len(df)):
    nl2.append(df.iloc[i,col2])
  nl2=list(set(nl2))
  print(nl2)
  val2=int(input("Enter the index of required item for event B\n\n"))
  c2=0
  for i in range(len(df)):
    if(df.iloc[i,col]==nl[val] and df.iloc[i,col2]==nl2[val2]):
      c2+=1
  panb=c2/(len(df)*2)
  #print(panb,c2)
  c3=0
  for i in range(len(df)):
    if(df.iloc[i,col2]==nl2[val2]):
      c3+=1
  pb=c3/len(df)
  pagb=panb/pb
  print("Conditional Probability of ",nl[val]," given ",nl2[val2]," is:\n",pagb)

  
```

```python

ch=int(input("Choose\n1. Conditional Probability\n2. Unconditional Probability\n"))
if(ch==1):
  con()
elif(ch==2):
  uncon()
else:
  print("Invalid choice\n")
```
